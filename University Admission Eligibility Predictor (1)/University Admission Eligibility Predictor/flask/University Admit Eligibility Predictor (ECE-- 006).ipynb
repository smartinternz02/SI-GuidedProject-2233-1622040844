# spliting data into train and test

```python
dataset.columns
```

```python
x= dataset.drop(columns=['Chance of Admit '])
```

```python
y=dataset['Chance of Admit ']
```

```python
x.shape
```

```python
y.shape
```

```python
x=np.array(x)
y=np.array(y)
```

```python
y=y.reshape(-1,1)
```

```python
y.shape
```

```python
from sklearn.preprocessing import StandardScaler
x_ss=StandardScaler()
X=x_ss.fit_transform(x)
```

```python
y_ss=StandardScaler()
y=y_ss.fit_transform(y)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=10)
```

```python
y_train=(y_train>0.5)
```

```python
y_test=(y_test>0.5)
```

```python
from sklearn.linear_model.logistic import LogisticRegression
cls=LogisticRegression(random_state=0)
lr=cls.fit(X_train,y_train)
```

```python
ypred=lr.predict(X_test)
```

```python
ypred
```

# Model Evaluation

```python
from sklearn.metrics import accuracy_score,recall_score,roc_auc_score,confusion_matrix
print("\nAccuracy score : %f" %(accuracy_score(y_test,ypred)*100))
print("Recall score : %f" %(recall_score(y_test,ypred)*100))
print("ROC score : %f\n" %(roc_auc_score(y_test,ypred)*100))
print(confusion_matrix(y_test,ypred))
```

# Saving the Model

```python
import pickle
pickle.dump(lr,open('university.pkl','wb'))
model=pickle.load(open('university.pkl','rb'))
```
